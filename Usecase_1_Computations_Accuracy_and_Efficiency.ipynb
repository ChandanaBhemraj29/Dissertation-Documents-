```python
#Code from Julius AI

import time

# Function to measure execution time
def measure_response_time(func, *args, **kwargs):
    # Record start time
    start_time = time.time()

    # Run the function
    result = func(*args, **kwargs)

    # Record end time
    end_time = time.time()

    # Calculate total execution time
    execution_time = end_time - start_time

    # Display the execution time
    print(f"Response time: {execution_time:.6f} seconds")

    return result

# Example of a function to test
def example_function(n):
    total = 0
    for i in range(n):
        total += i
    return total
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('HoC-GE2024-results-by-constituency.csv', encoding='utf-8')

# Calculate the total votes for each constituency
df['Total_Votes'] = df['Valid votes'] + df['Invalid votes']

# Calculate the percentage of total votes for each constituency
total_votes = df['Total_Votes'].sum()
df['Vote_Percentage'] = df['Total_Votes'] / total_votes * 100

# Sort constituencies by vote percentage in descending order
top_constituencies = df.sort_values('Vote_Percentage', ascending=False).head(10)

# Sum the votes for each party across all constituencies
party_votes = df[['Con', 'Lab', 'LD', 'RUK', 'Green', 'SNP', 'PC', 'DUP', 'SF', 'SDLP', 'UUP', 'APNI', 'All other candidates']].sum()

# Plotting the pie chart
plt.figure(figsize=(9, 5))
plt.pie(party_votes, labels=party_votes.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Share of National Votes by Political Party in UK Elections 2024')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()



#plt.show()

# Print the data for these top constituencies
#print(top_constituencies[['Constituency name', 'Vote_Percentage']])

# Measure the response time of the example function
measure_response_time(example_function, 1000000)
```
