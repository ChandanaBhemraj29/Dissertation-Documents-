```python
#Manual Visualisation Code

import time

# Function to measure execution time
def measure_response_time(func, *args, **kwargs):
    # Record start time
    start_time = time.time()

    # Run the function
    result = func(*args, **kwargs)

    # Record end time
    end_time = time.time()

    # Calculate total execution time
    execution_time = end_time - start_time

    # Display the execution time
    print(f"Response time: {execution_time:.6f} seconds")

    return result

# Example of a function to test
def example_function(n):
    total = 0
    for i in range(n):
        total += i
    return total





import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load the election results
election_data = pd.read_csv('/content/HoC-GE2024-results-by-constituency.csv')

# Load the GeoJSON file for constituency boundaries
geojson_data = gpd.read_file('/content/GeoJSON File.geojson')

# Merge the election data with the GeoJSON boundaries
merged_data = geojson_data.merge(election_data, left_on='PCON24NM', right_on='Constituency name')

# Define party vote columns
party_columns = ['Con', 'Lab', 'LD', 'Green', 'SNP', 'PC', 'DUP', 'SF', 'SDLP', 'UUP', 'APNI']

# Find the winning party for each constituency
merged_data['Winning Party'] = merged_data[party_columns].idxmax(axis=1)

# Define colors for each party
party_colors = {
    'Con': '#0087DC',  # Conservative
    'Lab': '#E4003B',  # Labour
    'LD': '#FAA61A',   # Liberal Democrats
    'Green': '#6AB023',# Green Party
    'SNP': '#FFF95D',  # Scottish National Party
    'PC': '#3F8428',   # Plaid Cymru
    'DUP': '#D46A4C',  # Democratic Unionist Party
    'SF': '#326760',   # Sinn Fein
    'SDLP': '#005B54', # Social Democratic and Labour Party
    'UUP': '#48A5EE',  # Ulster Unionist Party
    'APNI': '#FFD700', # Alliance Party of Northern Ireland
}

# Map winning party to color
merged_data['color'] = merged_data['Winning Party'].map(party_colors)

# Count total wins for each party
party_wins = merged_data['Winning Party'].value_counts()

# Plot the hex map with thinner black borders
fig, ax = plt.subplots(figsize=(15, 10))
merged_data.plot(ax=ax, color=merged_data['color'], edgecolor='white', linewidth=0.1)  # Adjust linewidth for border thickness

# Customize the plot
plt.title('2024 UK General Election Hex Map', fontsize=15)
plt.axis('off')

# Add legend with total wins for each party, positioned in the top right corner outside the map
handles = [plt.Line2D([0], [0], marker='o', color='w', label=f"{party} ({party_wins.get(party, 0)})", markersize=10, markerfacecolor=color)
           for party, color in party_colors.items()]
ax.legend(handles=handles, title='Winning Party (Total Wins)', loc='upper left', bbox_to_anchor=(1, 1), fontsize=9)

plt.show()



# Measure the response time of the example function
measure_response_time(example_function, 1000000)
```
